## Analisis de los postulantes

```python
postulantes_educacion = pd.read_csv('data/fiuba_1_postulantes_educacion.csv')
postulantes_educacion.head()
```

```python
postulantes_genero_edad = pd.read_csv('data/fiuba_2_postulantes_genero_y_edad.csv')
postulantes_genero_edad.head()
```

### Junto los dos DF's de postulantes

```python
postulantes = postulantes_educacion.join(postulantes_genero_edad.set_index('idpostulante'), on='idpostulante')
postulantes.head()
```

```python
postulantes.dtypes
```

```python
postulantes.sort_values('fechanacimiento')
```

```python
# Cantidad de postulantes
len(postulantes)
```

```python
postulantes.describe()
```

```python
# Distintos tipos de educacion
postulantes['nombre'].unique()
```

```python
grouped_by_education = postulantes.groupby(['nombre', 'estado'])
grouped_by_education.describe()
```

```python
grouped_by_status = postulantes.groupby('estado')
grouped_by_status.describe()
```
