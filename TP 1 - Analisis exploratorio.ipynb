## Analisis de los postulantes

```python
postulantes_educacion = pd.read_csv('data/fiuba_1_postulantes_educacion.csv')
postulantes_educacion.head()
```

```python
postulantes_genero_edad = pd.read_csv('data/fiuba_2_postulantes_genero_y_edad.csv')
postulantes_genero_edad.head()
```

### Junto los dos DF's de postulantes

```python
postulantes = postulantes_educacion.join(postulantes_genero_edad.set_index('idpostulante'), on='idpostulante')
postulantes.head()
```

```python
postulantes.dtypes
```

```python
# Convierto la fila de fecha de nacimiento a datetime
postulantes.fechanacimiento.apply(pd.to_datetime(postulantes.fechanacimiento, infer_datetime_format=True))
```

```python
# Cantidad de postulantes
len(postulantes)
```

```python
postulantes.describe()
```

```python
grouped_by_status = postulantes.groupby('estado')
grouped_by_status.describe()
```

## Analisis de los archivos de vistas/postulaciones

```python
postulantes_vistas = pd.read_csv('data/fiuba_3_vistas.csv')
postulantes_vistas.head()
```

```python
## Renombro al fila idAviso a idaviso para matchear el DF de abajo.
postulantes_vistas['idaviso'] = postulantes_vistas.idAviso
postulantes_vistas = postulantes_vistas.loc[:, ['idaviso','timestamp', 'idpostulante']]
postulantes_vistas.head()
```

```python
postulaciones = pd.read_csv('data/fiuba_4_postulaciones.csv')
postulaciones.head()
```

```python
vistas_postulaciones = postulantes_vistas.join(postulaciones.set_index('idaviso'), on='idaviso')
vistas_postulaciones.head()
```
